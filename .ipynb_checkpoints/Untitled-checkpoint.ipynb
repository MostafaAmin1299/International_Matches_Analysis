```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
main_df = pd.read_csv('international_matches.csv')
main_df.head()
```

```python
main_df.info()
```

```python
main_df.describe()
```

```python
df = pd.DataFrame(
            data=main_df
        )

df.head(10)
```

```python
df['h_diff'] = df['home_team_score'] - df['away_team_score']
df['a_diff'] = df['h_diff'] * -1

df.head(10)
```

```python
df['home_location'] = 'in'
df['away_location'] = 'in'

df.loc[df['neutral_location'] == True, ['home_location', 'away_location']] = 'neutral'

df.loc[df['home_team'] == df['country'], 'away_location'] = 'out'
df.loc[df['away_team'] == df['country'], 'home_location'] = 'out'

df
```

```python
for i in range(df.shape[0]):
    df = df.append({
                    'date': df['date'].iloc[i],
                    'home_team': df['away_team'].iloc[i],
                    'home_team_continent': df['away_team_continent'].iloc[i],
                    'home_team_fifa_rank': df['away_team_fifa_rank'].iloc[i],
                    'home_team_total_fifa_points': df['away_team_total_fifa_points'].iloc[i],
                    'h_diff':df['a_diff'].iloc[i],
                    'tournament': df['tournament'].iloc[i],
                    'city': df['city'].iloc[i],
                    'country': df['country'].iloc[i],
                    'home_location': df['away_location'].iloc[i],
                    'home_team_goalkeeper_score': df['away_team_goalkeeper_score'].iloc[i],
                    'home_team_mean_defense_score': df['away_team_mean_defense_score'].iloc[i],
                    'home_team_mean_offense_score': df['away_team_mean_offense_score'].iloc[i],
                    'home_team_mean_midfield_score': df['away_team_mean_midfield_score'].iloc[i],
                    'home_location': df['away_location'].iloc[i],
                }, 
                ignore_index=True
            )

df.tail(10)
```

```python
df.drop([
        'away_team', 
        'away_team_continent',
        'away_team_fifa_rank',
        'away_team_total_fifa_points',
        'home_team_score', 
        'away_team_score', 
        'a_diff', 
        'away_location', 
        'neutral_location',
        'shoot_out',
        'home_team_result',
        'away_team_goalkeeper_score',
        'away_team_mean_defense_score',
        'away_team_mean_offense_score',
        'away_team_mean_midfield_score'
    ], 
    axis = 1, inplace = True
)
df
```

```python
df.rename(
    columns={
        'home_team': 'team',
        'home_team_continent': 'team_continent',
        'h_diff': 'goal_difference',
        'home_team_fifa_rank': 'team_fifa_rank',
        'home_team_total_fifa_points': 'team_total_fifa_points',
        'city': 'match_city',
        'country': 'match_country',
        'home_location': 'match_in_country',
        'home_team_goalkeeper_score': 'team_goalkeeper_score',
        'home_team_mean_defense_score': 'team_mean_defense_score',
        'home_team_mean_offense_score': 'team_mean_offense_score',
        'home_team_mean_midfield_score': 'team_mean_midfield_score'
    },
    inplace=True, 
    errors='raise'
)

df.head()
```

```python
df.to_csv("data_after_clean.csv",index=False)
```
